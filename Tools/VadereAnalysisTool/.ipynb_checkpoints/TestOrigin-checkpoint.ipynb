```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import vadere_analysis_tool as v
from vadere_analysis_tool import OriginDeviation
import os
```

```python
p = v.VadereProject('../ContinuousIntegration/run_orign_translation_test.d')
```

```python
o2 = p.named_output.bridge_coordinates_kai_navigation_random_pos_with_offset()
o1 = p.named_output.bridge_coordinates_kai_navigation_random_pos_without_offset()
```

```python
test_orig = v.OriginDeviation([o1.output_dir, o2.output_dir])
```

```python
def get_output_pairs(path="../ContinuousIntegration/run_orign_translation_test.d'"):
    _output_pairs = []
    output_without_offset = [d for d in os.listdir(path) if d.endswith("without_offset")]
    for dir_without in output_without_offset:
        dir_with_offset = dir_without.replace("without_offset", "with_offset")
        name = dir_with_offset[:-len("without_offset")]
        pair = [os.path.join(path, dir_without), os.path.join(path, dir_with_offset)]
        _output_pairs.append(OriginDeviation(pair, name))

    return _output_pairs
```

```python
pairs = OriginDeviation.create_origin_deviation_pairs_from_project('../ContinuousIntegration/run_orign_translation_test.d')
results = [pair.get_origin_deviation_result() for pair in pairs]
```
